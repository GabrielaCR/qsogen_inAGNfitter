# The Richards+06 Big Blue Bump template as used in AGNfitter


```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import ascii
from astropy.table import Table, Column, MaskedColumn


path= '/Users/gcalistr/Documents/AGNfitter/'
R06dict = pickle.load(open(path + 'models/BBB/R06.pickle', 'rb'), encoding='latin1') 
bbb_nu, bbb_Fnu = R06dict['wavelength'], R06dict['SED'].squeeze()
renormfactor = 5e-53 #arbitraty renorm factor
plt.plot(bbb_nu, bbb_Fnu*10**bbb_nu * renormfactor ) #nuFnu , which is equivalent to lambda*F_lambda

plt.xlabel('log Rest frequency [Hz]')
plt.ylabel(r'$\lambda$ F$_{\lambda}$ ')
plt.yscale('log')
plt.ylim(1e-8,4e-6)
plt.xlim(13,17)
```

# Adding the Temple+21 SED template to compare

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import ascii
from astropy.table import Table, Column, MaskedColumn


from qsosed import Quasar_sed
import matplotlib.pyplot as plt

path= '/Users/gcalistr/Documents/AGNfitter/'
R06dict = pickle.load(open(path + 'models/BBB/R06.pickle', 'rb'), encoding='latin1') 
bbb_nu, bbb_Fnu = R06dict['wavelength'], R06dict['SED'].squeeze()


renormfactor = 5e-53 #arbitraty renorm factor
plt.plot(bbb_nu, bbb_Fnu*10**bbb_nu *renormfactor)


Quasar0 = Quasar_sed(z=0)
Quasar0.freq = 2.99792458e8/(Quasar0.wavlen*1e-10)
Quasar0.logfreq = np.log10(2.99792458e8/(Quasar0.wavlen*1e-10))
plt.plot(Quasar0.logfreq, Quasar0.wavlen*Quasar0.flux)


plt.xlabel('log Rest frequency [Hz]')
plt.ylabel(r'$\lambda$ F$_{\lambda}$ ')
plt.yscale('log')
plt.ylim(2e-8,1e-5)
plt.xlim(13,17)
```

# Temple+21 with add_blackbody() to compare

```python
path= '/Users/gcalistr/Documents/AGNfitter/'
R06dict = pickle.load(open(path + 'models/BBB/R06.pickle', 'rb'), encoding='latin1') 
bbb_nu, bbb_Fnu = R06dict['wavelength'], R06dict['SED'].squeeze()


renormfactor = 5e-53  #arbitraty renorm factor
plt.plot(bbb_nu, bbb_Fnu*10**bbb_nu *renormfactor)


Quasar0 = Quasar_sed(z=0)
Quasar0.add_blackbody()
Quasar0.freq = 2.99792458e8/(Quasar0.wavlen*1e-10)
Quasar0.logfreq = np.log10(2.99792458e8/(Quasar0.wavlen*1e-10))

plt.plot(Quasar0.logfreq, Quasar0.wavlen*Quasar0.flux)


plt.xlabel('Rest Wavelength [A]')
plt.ylabel('Flux density per unit wavelength')
plt.yscale('log')
plt.ylim(1e-8,4e-2)
plt.xlim(13,17)
```
